```python
#Importing Libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the data file
data = pd.read_csv("data_to_show.csv")
```

```python
#Creating input data and labels
X = data['x1']
y = data['y']
```

```python
#Converting the dataset into array and reshaping it
X=np.array(X).reshape(-1,1)
y=np.array(y).reshape(-1,1)
```

```python
#Creating and fitting LinearRegression model
model_1 = LinearRegression()
model_1.fit(X,y)
```

```python
#Predicting y for X using linear regression model
y_hat = model_1.predict(X)
```

```python
#Plotting the actual datapoints and the prediction results of the linear model
plt.plot(X, y, 'cx')
plt.plot(X, y_hat, 'b-')
plt.xlabel('x1')
plt.ylabel('y')
plt.title("Simpson's Paradox for existing data")
plt.show()
```

## Generating synthetic data to show Simpson's Paradox

```python
#Generating new synthetic data to see Simpson's Paradox Effect in that synthetic data
synthetic_X = []
synthetic_y = []
for i in range(3):
    #Generating 3000 different normally distributed values for x1 and x2
    x1 = np.random.normal(1, 10, 3000)
    x2 = np.random.normal(1, 10, 3000)
    #Generating a new constant for every new cluster
    constant = np.random.normal(1)
    for j in range(len(x1)):
        #Appending the values of y which is given by the equation y = 0.5*x1 + x2 + constant term
        #Here, I am adding and subtracting a certain number (30 and 60 respectively)
        #This is to show the progression and descent in the cluster patters as needed to depict the Simpson's Paradox.
        synthetic_X.append((0.8*x1[j]) + (30*i))
        synthetic_y.append((0.5*x1[j]) + x2[j] + constant - (60*i))
```

```python
#Converting the synthetic dataset into array and reshaping it
data_X = np.array(synthetic_X).reshape(-1, 1)
data_y = np.array(synthetic_y).reshape(-1, 1)
```

```python
#Creating and fitting LinearRegression model
model_2 = LinearRegression()
model_2.fit(data_X,data_y)
```

```python
#Predicting y for synthetic X using linear regression model
y_pred = model_2.predict(data_X)
```

```python
#Plotting the actual datapoints and the prediction results of the linear model
plt.plot(data_X, data_y, 'gx')
plt.plot(data_X, y_pred, 'r-')
plt.xlabel('X')
plt.ylabel('y')
plt.title("Simpson's Paradox for synthetic data")
plt.show()
```
